```python
X = data.drop(columns = ['timestamp','t2',
                        'is_holiday','is_weekend',
                        'season','year','month','day','cnt'])

y = data['cnt']
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.25)
```

```python
lr = LinearRegression(fit_intercept = True, normalize = True)
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

```python
# round(y_pred.apply(lambda x:))
y_pred=list(map((lambda x: round(x)),y_pred))
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error #
from sklearn.metrics import median_absolute_error, r2_score
```

```python
print("Средняя абсолютная ошибка: {:.3f}".format(mean_absolute_error(y_test, y_pred)))
print("Средняя квадратичная ошибка: {:.3f}".format(mean_squared_error(y_test, y_pred, squared = False)))
print('Медианное абсолютное отклонение: {:.3f}'.format(median_absolute_error(y_test, y_pred)))
print('Коэффициент детерминации: {:.3f}'.format(r2_score(y_test, y_pred)))
```

```python
def graf(col_name):
    plt.figure(figsize=(15, 11))
    x_low = 0.9*min(X_test[col_name])
    x_high = 1.1*max(X_test[col_name])
    x_extended = np.linspace(x_low, x_high, 100)
    line = np.polyfit(X_test[col_name], y_pred, 1) 
    plt.plot(x_extended, np.polyval(line,x_extended),'r-', lw=3)
    plt.xlim(x_low, x_high)
    plt.xlabel(col_name)
    plt.ylabel('Кол-во велосипедов')
    plt.scatter(X_test[col_name], y_test,  color='blue')
    plt.legend(('Найденная линейная зависимость', 'Данные'), loc='best')
    plt.title('Предсказание кол-ва велосипедов')
    plt.show()
```

```python
graf('t1')
# timestamp        object
# cnt               int64
# t1              float64
# t2              float64
# hum             float64
# wind_speed      float64
# weather_code    float64
# is_holiday      float64
# is_weekend      float64
# season          float64
# year              int64
# month             int64
# day               int64
# hour              int64
# dtype: object
```

```python
X = data.drop(columns = ['timestamp',
                        'is_holiday','is_weekend',
                        'season','year','month','day','cnt'])
y = data['cnt']
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.25)

lr = LinearRegression(fit_intercept = True, normalize = True)
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)

y_pred=list(map((lambda x: round(x)),y_pred))
print('Зависимые параметры: время, температура воздуха, ощущаемая температура, влажность воздуха, код погоды')
print("Средняя абсолютная ошибка: {:.3f}".format(mean_absolute_error(y_test, y_pred)))
print("Средняя квадратичная ошибка: {:.3f}".format(mean_squared_error(y_test, y_pred, squared = False)))
print('Медианное абсолютное отклонение: {:.3f}'.format(median_absolute_error(y_test, y_pred)))
print('Коэффициент детерминации: {:.3f}'.format(r2_score(y_test, y_pred)))
```
